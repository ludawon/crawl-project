```python
import GMethods 
import pandas as pd
```

```python
driver = GMethods.getChrome(mode=1)
driver.get('https://www.worldometers.info/coronavirus/country/indonesia/')
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(driver.page_source,'lxml')
```

```python
charts = soup.find_all('div',{'class':'row graph_row'})
charts[0].find('script')
```

```python
date = driver.execute_script('return window.Highcharts.charts[0]''.xAxis[0].categories')
total_case = driver.execute_script('return window.Highcharts.charts[0]''.series[0].options.data')
print(total_case)
```

```python
date = driver.execute_script('return window.Highcharts.charts[3]''.xAxis[0].categories')
active = driver.execute_script('return window.Highcharts.charts[3]''.series[0].options.data')
for i in range(len(date)):
    print(date[i],active[i],total_case[i])
```

```python
date = driver.execute_script('return window.Highcharts.charts[5]''.xAxis[0].categories')
deaths = driver.execute_script('return window.Highcharts.charts[5]''.series[0].options.data')
for i in range(len(date)):
    print(date[i],deaths[i],total_case[i])
```

```python
df = pd.DataFrame({
    'date' : date,
    'case' : total_case,
    'active' : active,
    'deaths' : deaths
    
})

day = list()
for i in range(len(date)):
    day.append(i+1)
df['Day'] = day

recovered = list()
for i in range(len(date)):
    recovered.append(df.case.loc[i]-df.active.loc[i]-df.deaths.loc[i])
df['recovered'] = recovered

new_case = list()
for i in range(len(date)):
    if(i==0):
        new_case.append(df.case.loc[i])
    else:
        new_case.append(df.case.loc[i]-df.case.loc[i-1])
df['daily_case'] = new_case

daily_recovered = list()
for i in range(len(date)):
    if(i==0):
        daily_recovered.append(df.recovered.loc[i])
    else:
        daily_recovered.append(df.recovered.loc[i]-df.recovered.loc[i-1])
df['daily_recovered'] = daily_recovered

daily_deaths = list()
for i in range(len(date)):
    if(i==0):
        daily_deaths.append(df.deaths.loc[i])
    else:
        daily_deaths.append(df.deaths.loc[i]-df.deaths.loc[i-1])
df['daily_deaths'] = daily_deaths

closed = list()
for i in range(len(date)):
    if(i==0):
        closed.append(df.case.loc[i])
    else:
        closed.append(df.case.loc[i]-df.active.loc[i])
df['closed'] = closed

active_rate = list()
closed_rate = list()
for i in range(len(date)):
    case = df.case.loc[i]
    if(case==0):
        active_rate.append(0)
        closed_rate.append(1)
    else:
        active_rate.append(df.active.loc[i]/df.case.loc[i])
        closed_rate.append(df.closed.loc[i]/df.case.loc[i])
df['active_rate'] = active_rate
df['closed_rate'] = closed_rate

three_avg = list()
for i in range(len(date)):
    if(i==0):
        three_avg.append(df.daily_case.loc[0])
    elif(i==1):
        three_avg.append((df.daily_case.loc[0]+df.daily_case.loc[1])/3)
    else:
        three_avg.append((df.daily_case.loc[i-2]+df.daily_case.loc[i-1]+df.daily_case.loc[i])/3)
df['3_days_avg'] = three_avg
    
seven_avg = list()
for i in range(len(date)):
    temp = 0
    if(i<6):
        for j in range(i):
            temp = temp+df.daily_case.loc[j]
        seven_avg.append(temp/7)
    else:
        for j in range(7):
            temp = temp+df.daily_case.loc[i-j]
        seven_avg.append(temp/7)
df['7_days_avg'] = seven_avg    

two_week_case = list()
daily_two_weeks_ago = list()
for i in range(len(date)):
    temp = 0
    if(i<14):
        two_week_case.append(0)
        daily_two_weeks_ago.append(0)
    else:
        twa = df.daily_case.loc[i-14]
        temp =  df.daily_recovered.loc[i] - twa
        daily_two_weeks_ago.append(twa)
        two_week_case.append(temp)
df['daily_two_weeks_ago'] = daily_two_weeks_ago
df['two_week_case'] = two_week_case    

delta_active_rate = list()
delta_closed_rate = list()
for i in range(len(date)):
    if(i==0):
        delta_active_rate.append(0)
        delta_closed_rate.append(0)
    else:
        delta_active_rate.append(df.active_rate.loc[i]-df.active_rate.loc[i-1])
        delta_closed_rate.append(df.closed_rate.loc[i]-df.closed_rate.loc[i-1])
df['delta_active_rate'] = delta_active_rate
df['delta_closed_rate'] = delta_closed_rate

df = df[['date','Day','case','active','active_rate','delta_active_rate',
         'closed','closed_rate','delta_closed_rate','recovered','deaths','daily_case',
         '3_days_avg','7_days_avg','daily_deaths','daily_recovered',
         'daily_two_weeks_ago','two_week_case']]
df
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
 
# Dataset
x=df['Day']
y=df['active']
 
cubic_interploation_model = interp1d(x, y, kind = "cubic")
 
# Plotting the Graph
X_=np.linspace(x.min(), x.max(), 500)
Y_=cubic_interploation_model(X_)

plt.figure(figsize=(12, 7), dpi=80) 
plt.plot(X_, Y_)
plt.title("Plot Smooth Curve Using the scipy.interpolate.interp1d Class",
          horizontalalignment="center",fontstyle = "normal", 
          fontsize = "22", fontfamily = "sans-serif")
plt.xlabel("Day")
plt.ylabel("Y")
plt.show()
```
